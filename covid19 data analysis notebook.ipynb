### task 3.1: caculating the first derivative of the curve

### task 3.2: find maxmimum infection rate for China

```python
corona_dataset_agregated.loc['China'][:60].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_agregated.index)
max_infection_rates = []
for c in countries :
    max_infection_rates.append(corona_dataset_agregated.loc[c].diff().max())
corona_dataset_agregated["max_infection_rates"] = max_infection_rates


```

```python
corona_dataset_agregated.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(corona_dataset_agregated["max_infection_rates"])
```

```python
corona_data.head()
```
